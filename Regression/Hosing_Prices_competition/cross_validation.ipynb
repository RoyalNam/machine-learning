```python
import pandas as pd
data = pd.read_csv('./test.csv', index_col='Id')
y = pd.read_csv('./sample_submission.csv').SalePrice
```

```python
data.head()
```

```python
numeric_col = [cname for cname in data.columns if data[cname].dtype in ['int64','float64']]
X = data[numeric_col].copy()
```

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor

my_pipeline = Pipeline(steps=[
    ('impute',SimpleImputer()),
    ('model',RandomForestRegressor(n_estimators=50, random_state=0))
])
```

```python
from sklearn.model_selection import cross_val_score
scores = -1* cross_val_score(my_pipeline, X,y,cv=5,scoring='neg_mean_absolute_error')
print('MEA scores: ',scores)
```

```python
print("Average MEA score: ",scores.mean())
```
