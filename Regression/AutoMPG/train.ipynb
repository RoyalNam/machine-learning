```python
import pandas as pd
import numpy as np

col_name = ['mpg','cylinders','displacement','horsepower','weight','acceleration','model year','origin','car name']

data = pd.read_csv('auto-mpg.data',  delimiter=r'\s+',names=col_name, index_col='car name')
data.replace('?', np.nan, inplace=True)
data.dropna(inplace=True)
data.head()
```

```python
data.info()
```

```python
data[['horsepower']]  = data[['horsepower']].astype(float)
```

```python
data.info()
```

```python
X = data.drop('mpg', axis=1)
y = data.mpg
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=0, test_size=0.2)
```

```python
from sklearn.metrics import mean_absolute_error
def getMEA(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    
    print(np.concatenate((pred.reshape(len(pred), 1), y_test.values.reshape(len(y_test), 1)),1))
    
    return mean_absolute_error(y_test, pred)
```

```python
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
print('MEA:', getMEA(model1, X_train, X_test, y_train, y_test))
```

```python
from sklearn.svm import SVR
model2 = SVR(kernel='rbf')
print('MEA:', getMEA(model2, X_train, X_test, y_train, y_test))
```

```python
from sklearn.tree import DecisionTreeRegressor
model3 = DecisionTreeRegressor(random_state=0)
print('MEA:', getMEA(model3, X_train, X_test, y_train, y_test))
```

```python
from sklearn.ensemble import RandomForestRegressor
model4 = RandomForestRegressor(n_estimators=40)
print('MEA:', getMEA(model4, X_train, X_test, y_train, y_test))
```
