```python
col_name = ['Sex', 'Length', 'Diameter', 'Height', 'Whole_weight', 'Shucked_weight', 'Viscera_weight', 'Shell_weight', 'Rings']
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
data = pd.read_csv('abalone.data', names=col_name)
data
```

```python
data.isnull().sum()
```

```python
X = data.drop('Rings', axis=1)
y = data.Rings
```

```python
from sklearn.preprocessing import OneHotEncoder
encode = OneHotEncoder(handle_unknown='ignore', sparse=False)
sex_transform_df = pd.DataFrame(encode.fit_transform(X[['Sex']]), columns=encode.get_feature_names_out(['Sex']))
sex_transform_df


X = pd.concat([X, sex_transform_df], axis=1)
X.drop('Sex', axis=1, inplace=True)
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.2)
```

```python
y_test
```

```python
X_train
```

```python
from sklearn.metrics import mean_absolute_error
def getMEA(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_pred = y_pred.round().astype(int)
    return mean_absolute_error(y_test, y_pred)
```

```python
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
print("MEA Random Fores:", getMEA(model1, X_train, X_test, y_train, y_test))
```

```python
from sklearn.svm import SVR
model2 = SVR(kernel='rbf')
print("MEA Random Fores:", getMEA(model2, X_train, X_test, y_train, y_test))
```

```python
from sklearn.tree import DecisionTreeRegressor
model3 = DecisionTreeRegressor()
print("MEA Random Fores:", getMEA(model3, X_train, X_test, y_train, y_test))
```

```python
from sklearn.ensemble import RandomForestRegressor
model4 = RandomForestRegressor(random_state=0)
print("MEA Random Fores:", getMEA(model4, X_train, X_test, y_train, y_test))
```
